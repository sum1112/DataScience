# Example 1 - A simple cost function

$f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

```python
x_1 = np.linspace(start=-3,stop=3,num=100)
print(x_1)
```

```python
plt.xlim([-3,3])
plt.ylim([0,12])
plt.xlabel('X', fontsize=16)
plt.ylabel('f (X)', fontsize=16)
plt.plot(x_1, f(x_1))
plt.show()
```

```python
def df(x):
    return 2 * x + 1
```

```python

plt.figure(figsize=[8,8])

plt.subplot(2, 1, 1)

plt.xlim([-3,3])
plt.ylim([0,12])
plt.title("Cost Function")
plt.xlabel('X', fontsize=16)
plt.ylabel('f (X)', fontsize=16)
plt.plot(x_1, f(x_1), color="blue", linewidth=3)

plt.subplot(2, 1, 2)

plt.grid()
plt.title("Slope of the Cost Function")
plt.xlabel("X")
plt.ylabel("df (X)")
plt.xlim([-3, 3])
plt.ylim([-9,6])

plt.plot(x_1, df(x_1), color="skyblue", linewidth=3)

plt.show()
```

# Python loops and gradient descent

```python
for x in range(5):
    print("Hello world ", x)
```

```python
# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = .1
precision = .0001
x_list = [new_x]
slope_list = [f(new_x)]

for n in range(50):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - gradient * step_multiplier
    
    step_size = abs(new_x - previous_x)
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    if step_size < precision:
        print("Loop ran this many times:", n)
        break
    
```

```python
print("Local Minimum at ", new_x)
print("Slope at this ", df(new_x))
print("cost is ", f(new_x))
```

```python

plt.figure(figsize=[10,5])

plt.subplot(1, 2, 1)

plt.xlim([-3,3])
plt.ylim([0,12])
plt.title("Cost Function")
plt.xlabel('X', fontsize=16)
plt.ylabel('f (X)', fontsize=16)
plt.plot(x_1, f(x_1), color="blue", linewidth=3)
values = np.array(x_list)
plt.scatter(values, f(values), color="red", s=100, alpha=0.6)

plt.subplot(1, 2, 2)

plt.grid()
plt.title("Slope of the Cost Function")
plt.xlabel("X")
plt.ylabel("df (X)")
plt.xlim([-3, 3])
plt.ylim([-9,6])

plt.plot(x_1, df(x_1), color="skyblue", linewidth=3)
values2 = np.array(slope_list)
plt.scatter(x_list, values2, color="red", s=100, alpha=0.6)

plt.show()
```
